```python
# Importing libraries
import pandas as pd
import warnings
import os 
import glob 
import numpy as np
from sqlalchemy import create_engine
import mysql.connector

import credentials
```

### Importing Data (Universities major's data information from excel files)

```python
#### Importing data
#### Copy and paste the route where all de excel files are

## from credential
##path = "D:\\Coreano\\gks24\\UniversitiesInformation2024\\UniverstyDataSets"
path = credentials.path


dir_list = os.listdir(path)
DF_list = list()
for x in dir_list:
    with warnings.catch_warnings():
        warnings.filterwarnings("ignore", category=UserWarning)
        ##df = pd.read_excel("file.xlsx") 
        df = pd.read_excel('UniverstyDataSets/'+x)
        DF_list.append(df)
        print(x)
```

## Step 2 Data cleaning 

### Getting the same columns for all the universites erasing non data columns, erasing missing rows and renaming the column names with number (20 total columns) and concating all the university files in just one dataframe

```python
## Data cleaining

i=0
for uniDF in DF_list:        
    column_name = "Unnamed: 1"
    try:
        ### df = df.drop(column_name, axis=1)
        ### Removing Empty columns
        uniDF=uniDF.dropna(subset=[column_name]).reset_index(drop=True)
        header_row = 0
        ### Renaming headers and erase the column with headers name
        uniDF.columns = uniDF.iloc[header_row]
        uniDF = uniDF.drop(uniDF[uniDF['No.'] == 'No.'].index)

    except KeyError:
        print(f"Column '{column_name}' not found.")
    
    uniDF.columns = uniDF.columns.fillna('EmptyColumn')
    columns_to_delete = ["모집 희망\n여부","Major","전공명","비고","EmptyColumn"]
    for col in columns_to_delete:
        try:
   #        ### df = df.drop(col, axis=1)
            ## Erase clumns that does not match with the rest or the universisites of the list
            #uniDF = uniDF.drop(np.nan, axis=1)
            #uniDF = uniDF.dropna(subset=col).reset_index(drop=True)
            uniDF = uniDF.drop(col, axis=1)
            uniDF = uniDF.loc[:, pd.notnull(uniDF.columns)]
            print(f"Column '{col}' FOUNDED.")
        except KeyError:
            print(f"Column '{col}' not")
            
    ## merge on a DF list
    DF_list[i] = uniDF
    i+=1

```

```python
for uniDF in DF_list:
    uniDF.columns = range(uniDF.columns.size)
```

```python
df_concat = pd.concat(DF_list)
```

### Conect to database

```python
mydb = mysql.connector.connect(
  host = credentials.mySQL_host,
  user = credentials.mySQL_user,
  password = credentials.mySQL_password,
  database = credentials.mySQL_database
)
```

```python
mycursor = mydb.cursor()
mycursor.execute(""" CREATE TABLE IF NOT EXISTS universities 
                 (Num int, 
                  대학명 VARCHAR(255), 
                  University_대학알리미_공시_명칭 VARCHAR(255), 
                  Embassy_Track_Type VARCHAR(255),
                  Univ_Track_Applicable_Programs VARCHAR(255),
                  Campus_Location VARCHAR(255),
                  학과명 VARCHAR(255),
                  Department_대학알리미_공시_명칭 VARCHAR(255),
                  학과계열 VARCHAR(255),
                  Field_of_Study_Division VARCHAR(255),
                  Masters VARCHAR(255),
                  Medium_of_Instruction  VARCHAR(255),
                  Required_TOPIK_for_Admission_after_Language_Program VARCHAR(255),
                  Program_starts VARCHAR(255),
                  Doctoral VARCHAR(255),
                  Medium_of_Instruction_doc VARCHAR(255),
                  Required_TOPIK_for_Admission_after_Language_Program_doc VARCHAR(255),
                  Program_starts_doc VARCHAR(255),
                  Website_URL_for_detailed_information VARCHAR(255),
                  Remarks VARCHAR(255) )
                  """)
```

```python
#Create a SQLAlchemy engine to connect to the MySQL database
engine = create_engine(credentials.engine_name)

# Step 3: Convert the Pandas DataFrame to a format for MySQL table insertion
finalResult.to_sql('universities', con=engine, if_exists='replace', index=False)
```
